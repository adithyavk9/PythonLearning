<h1 style="color:#6C3483" align="center">Python: Programming Environment</h1>

<h2 align="left">1. Characteristics of Python</h2>
<p style="font-size:17px">
a) Easy to learn, powerful programming language<br>
b) Efficient high-level data structures <br>
c) Simple but effective approach to object-oriented programming<br>
d) Extensive standard libraries: freely available and open source<br>
e) Easy integration of C and C++ codes
</p>

<h2 align="left">2. Installing Python</h2>
<h3 align="left">2.1 Choosing Suitable Python Version</h3>
<p style="font-size:17px">
a) Python 2.7 (https://www.python.org/)<br>
b) Python 3.# (https://www.python.org/)<br>
c) IPython (https://ipython.org/): Not a version but it has completion, history, shell capabilities, and a lot more.
</p>
<h3 align="left">2.2 Popular Python Editors</h3>
<p style="font-size:17px">
a) Jupyter Notebook (https://jupyter.org/)<br>
b) Spyder (https://spyder-ide.github.io/)<br>
c) PyCharm (https://www.jetbrains.com/pycharm/)
</p>
<h3 align="left">2.3 A Compact Python Distribution</h3>
<p style="font-size:17px">
a) Anaconda Distribution: IDE, Package Manager (Conda), and Python Interpreter (https://www.anaconda.com/)
</p>
<h3 align="left">2.4 Package Installation</h3>
<p style="font-size:17px">
Run the following commands in Terminal (Linux) or in command prompt (Windows):<br>
a) <b>conda install 'package_name'</b> (for anaconda distribution) <br>
b) <b>pip install 'package_name'</b> (for python 2.7 distribution)<br>
c) <b>pip3 install 'package_name'</b> (for python 3.# distribution)<br>
</p>

<h2 align="left">3. Pillars of Python: Packages</h2>
<p style="font-size:17px">
a) <b>NumPy:</b> Numerical computing and routines with powerful numerical arrays objects <br>
b) <b>SciPy:</b> High-level numerical routines, optimization, regression, interpolation, etc  <br>
c) <b>Matplotlib / Mayavi (for  3D):</b> Data visualization (plotting) <br>
d) <b>Pillow / Scikit-image / OpenCV:</b> Image processing <br>
e) <b>SciKit-learn:</b> Machine learning <br>
f) <b>Pandas / Statsmodels / Seaborn:</b> Statistics <br>
g) <b>SymPy:</b> Symbolic computing - algebric maniuplations, calculus, solving equations, etc <br>
h) <b>Pytorch / Tensorflow / Keras:</b> Deep learning <br>
i) <b>wxPython / pyQT:</b> GUI toolkit
</p>
