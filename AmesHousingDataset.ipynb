```python
plt.figure(figsize=(15, 10))

plt.subplot(2,2,1)
plt.scatter(df.LotArea , df.SalePrice, c = "blue", marker = ".")
plt.title("Lot Area vs Sale Price")
plt.xlabel("Lot Area")
plt.ylabel("Sale Price")

plt.subplot(2,2,2)
plt.scatter(df.BedroomAbvGr , df.SalePrice, c = "green", marker = "o" )
plt.title("Bedroom vs Sale Price")
plt.xlabel("Bedroom Abv Gr")
plt.ylabel("Sale Price")

plt.subplot(2,2,3)
plt.scatter(df.LotArea , df.SalePrice, c = "blue", marker = ".")
plt.title("Lot Area vs Sale Price")
plt.xlabel("Lot Area")
plt.ylabel("Sale Price")

plt.show()
```
