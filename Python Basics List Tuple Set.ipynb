Problem: Calculating the sum of numbers in a list using a for loop

```python
# Define a list of numbers
numbers = [1, 2, 3, 4, 5]

# Initialize a variable to store the sum
sum_of_numbers = 0

# Use a for loop to iterate over the numbers in the list
for num in numbers:
    # Add the current number to the sum
    sum_of_numbers += num

# Print the result
print(f"The sum of numbers is: {sum_of_numbers}")
```

```python
##### Tuple ###############
#1. single variable can store multiple data.
#2. It is also ordered.
#3. It is not mutable.
#4. Normally faster than a list

t = (23, 20, 6, 5,20, 6) # tuple is created using the round braces (during creation, we should be careful about the braces).
print(t)

print(t[2]) # Indexing is supported so it is ordered. For accessing, always use square braces.
print(type(t)) # type is a library function so round braces.
```

```python
t[2] = 10   # tuples are not mutable or changable.
```

```python
print(t.count(20)) # count is used to count the number occurances of a value.
print(t.index(6)) # index returns the first occurance of a value
```

```python
## Changing Tuple Values

stationeries = ("pen", "pencil", "eraser", "marker")
print("initial tuple: ", stationeries)
```

```python
list_of_stationeries = list(stationeries)
list_of_stationeries[1] = "correction pen"
stationeries = tuple(list_of_stationeries)
print("modified tuple: ", stationeries)
```

```python
## Add items

# 1. Convert into a list

fruits_tuple = ("apple", "banana", "cherry")
print("initial tuple: ", fruits_tuple)
list_of_fruits = list(fruits_tuple)
list_of_fruits.append("orange")
fruits_tuple = tuple(list_of_fruits)
print("modified tuple: ", fruits_tuple)
```

```python
# 2. Add tuple to a tuple

fruits_tuple = ("apple", "banana", "cherry")
print("initial tuple: ", fruits_tuple)
orange = ("orange",)
fruits_tuple += orange  # we can also write: fruits_tuple = fruits_tuple + orange
print("modified tuple: ", fruits_tuple)
```

<h3 align="left">Set: No duplicates</h3>

```python
#### Set #########
#1. A set is collection of unique elemnts.
#2. for creating a set we will use a curley braces.
#3. Set is not ordered.

'''
However, there are two important restrictions due to the algorithmic underpinnings:
1. The first is that the set does not maintain the elements in any particular order.
2. The second is that only instances of immutable types can be added to a Python set.

Therefore, objects such as integers, floating-point numbers,
and character strings are eligible to be elements of a set.
It is possible to maintain a set of tuples, but not a set
of lists or a set of sets, as lists and sets are mutable.
The frozenset class is an immutable form of the set type,
so it is legal to have a set of frozensets.
'''

s = {2,3,1,8,5,96,2} #  duplicates are not allowed in a set
s1 = {2,4,5,"palash", '4'}
print(type(s))
print(s, s1)
```

```python
print(s1.intersection(s))
```

```python
s1.add(3)
s1.add(2+3j)
print(s1) #Indexing is not allowed because it does not have any order.
```

#### Set Methods

1. add(): adds an element to the set
2. clear(): removes all elements in a set
3. copy(): copies the set
4. difference(): returns a set that contains the difference between two sets
5. difference_update(): removes the items that exist in both sets
6. discard(): removes the specified item from the set
7. intersection(): returns a set that contains the similarity between two or more sets
8. intersection_update(): removes the items that is not present in both sets (or in all sets if the comparison is done between more than two sets)
9. isdisjoint(): returns True if none of the items are present in both sets, otherwise it returns False
10. issubset(): returns True if all items in the set exists in the specified set, otherwise it retuns False
11. issuperset(): returns True if all items in the specified set exists in the original set, otherwise it retuns False
12. pop(): removes a random item from the set
13. remove(): removes the specified element from the set
14. symmetric_difference(): returns a set that contains all items from both set, but not the items that are present in both sets
15. symmetric_difference_update(): updates the original set by removing items that are present in both sets, and inserting the other items
16. union(): returns a set that contains all items from the original set, and all items from the specified set(s)
17. update(): updates the current set, by adding items from another set (or any other iterable)

#### Dictionary Methods


1. clear(): Removes all the elements from the dictionary
2. copy(): Returns a copy of the dictionary
3. fromkeys(): Returns a dictionary with the specified keys and value
4. get(): Returns the value of the specified key
5. items(): Returns a list containing a tuple for each key value pair
6. keys(): Returns a list containing the dictionary's keys
7. pop(): Removes the element with the specified key
8. popitem(): Removes the last inserted key-value pair
9. setdefault(): Returns the value of the specified key. If the key does not exist: insert the key, with the specified value
9. update(): Updates the dictionary with the specified key-value pairs
10. values(): Returns a list of all the values in the dictionary

```python
## Write a Python program to Extract Unique dictionary values.

my_dict = {
    'a': 10,
    'b': 40,
    'c': 10,
    'd': 40,
    'e': 30
}

uni_vals = set()

for i in my_dict.values():
  uni_vals.add(i)

unique_values_list = list(uni_vals)
print(unique_values_list)
```
