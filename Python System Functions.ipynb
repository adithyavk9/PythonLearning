### 1. Import

```python
import os
import shutil
```

### 2. Knowing present working directory

```python
# Knowing present working directory
xx = os.getcwd()
print(xx)
```

### 3. Listing files and folders present in the directory

```python
# Knowing files and folders present in the directory

yy = os.listdir(xx)
print(yy)

for file_folder in yy:
    print(file_folder)
```

### 4. Change the present working directory

```python
# Change the directory

path = "./Python Prog for Data Eng"

os.chdir(path)
print(os.getcwd())
```

```python
os.chdir(xx)
print(os.getcwd())
```

### 5. Creating new directory

```python
# Creating new directory

dirName = xx + '/test'

# Checking existance of folder
if not os.path.isdir(dirName):
    print('The directory is being created')
    os.mkdir(dirName)

print(dirName)
```

### 6. Creating nested (or chain of) directories

```python
# Creating chain of directories

dirName_new = dirName + '/test_1/test_2/test_3/test_4/'

os.makedirs(dirName_new)
```

### 7. Checking the existance of a file

```python
# Checking existance of file

fileName = xx + '/4.json'

if os.path.isfile(fileName):
    print(fileName)
```

### 8. Copying the files/folders

```python
# Copying of file/folder

sourceFile = fileName
destFile = dirName + '/test_1/test_2/test_3/4.json'

shutil.copy(sourceFile, destFile)

# Copy and rename

sourceFile = dirName + '/test_1/test_2/test_3/4.json'
destFile = dirName + '/test_1/a_1_file.json'

shutil.copy(sourceFile, destFile)
```

### 9. Moving the files/folders

```python
# Moving of file/folder

sourceFile = dirName + '/test_1/a_1_file.json'
destFile = dirName + '/test_1/test_2/a_1_file.txt'

shutil.move(sourceFile, destFile)
```

### 10. Renaming file/folder

```python
# Renaming file
oldName = xx + '/' + 'brics.xlsx'
newName = xx + '/' + 'N_brics.xlsx'

os.rename(oldName,newName)
```
