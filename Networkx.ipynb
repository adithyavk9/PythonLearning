How to create graphs using networkx?


```python
G = nx.Graph()  #empty graph object created
G.add_nodes_from([1,2,3]) #to add multiple nodes add_nodes_from is used
G.add_edges_from([(1,2),(2,3),(1,1)]) #to add multiple edges add_edges_from is used
```

```python
G.add_node(4) #to add a single node
G.add_edge(1,4) #to add a single edge
print(G.nodes())  #to print all the nodes in G
print(G.edges())
```

```python
for u,v in G.edges():
  if u==v:
    G.remove_edge(u,v)
print(G.edges())
```

```python
G.add_edge(1,3,weight=4.2) #adding edges with weight
print(G.edges(data=True)) #to print edges with weight
```

Downloading a social network from internet and creating a graph.

```python
import networkx as nx
import matplotlib.pyplot as plt
import urllib.request
import gzip
import shutil
```

```python
url = 'https://snap.stanford.edu/data/ca-HepTh.txt.gz'
dataset_path = 'ca-HelpTh.txt.gz'
extracted_path = 'ca-HelpTh.txt'

urllib.request.urlretrieve(url,dataset_path)

with gzip.open(dataset_path,'rb') as f_in:
  with open(extracted_path,'wb') as f_out:
    shutil.copyfileobj(f_in,f_out)
#load the data set to networkx graph
G = nx.read_edgelist(extracted_path,comments='#',nodetype = int)
```

```python
#node count , edge count and average degree
node_count = G.number_of_nodes()
edge_count = G.number_of_edges()
average_degree = sum(dict(G.degree()).values())/float(node_count)
print(f'Node Count: {node_count}')
print(f'Edge Count: {edge_count}')
print(f'Average Degree: {average_degree:.2f}')
```
